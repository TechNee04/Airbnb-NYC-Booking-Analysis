```python
# Cap the outliers in the 'price' column using IQR technique
q1 =  Airbnb_df['price'].quantile(0.25)
q3 =  Airbnb_df['price'].quantile(0.75)
iqr = q3 - q1
upper_threshold = q3 + 1.5 * iqr
lower_threshold = q1 - 1.5 * iqr
Airbnb_df['price'] = Airbnb_df['price'].clip(lower=lower_threshold, upper=upper_threshold)
```
